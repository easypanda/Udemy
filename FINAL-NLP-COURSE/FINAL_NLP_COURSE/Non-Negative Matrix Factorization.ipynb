```python
import pandas as pd
import numpy as np
```

```python
npr = pd.read_csv("./npr.csv")
```

```python
npr.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_df=0.95,min_df=2,stop_words="english") #min_df means filter words appear unique documents
```

```python
dtm = tfidf.fit_transform(npr["Article"])
```

```python
dtm
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(n_components=7,random_state=42)
```

```python
nmf_model.fit(dtm)
```

```python
tfidf.get_feature_names()[2300]
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f"THE TOP 15 WORDS FOR TOPIC #{index}")
    print([tfidf.get_feature_names()[i] for i in topic.argsort()[-15:]])
    print("\n")
```

```python
topic_results = nmf_model.transform(dtm)
```

```python
topic_results[0].round(2).argmax()
```

```python
topic_results.argmax(axis=1)
```

```python
npr["topic"] = topic_results.argmax(axis=1)
```

```python
npr.head()
```

```python
dict_topic = {0:"health",1:"election",2:"legas",3:"politics",4:"election",5: "music", 6:"edu"}
```

```python
npr["Topic_name"] = npr["topic"].map(dict_topic)
```
