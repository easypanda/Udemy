## Working with PDF Files
#### 4. Use PyPDF2 to open the file `Business_Proposal.pdf`. Extract the text of page 2.

```python
# Perform import
import PyPDF2

# Open the file as a binary object
file = open("Business_Proposal.pdf","rb")

# Use PyPDF2 to read the text of the file
mypdf = PyPDF2.PdfFileReader(file)


# Get the text from page 2 (CHALLENGE: Do this in one step!)
page_two_text = mypdf.getPage(1).extractText()



# Close the file

file.close()
# Print the contents of page_two_text
print(page_two_text)
```

#### 5. Open the file `contacts.txt` in append mode. Add the text of page 2 from above to `contacts.txt`.

#### CHALLENGE: See if you can remove the word "AUTHORS:"

```python
# Simple Solution:
my_file = open('contacts.txt','a+')
my_file.write(page_two_text)
my_file.seek(0)
print(my_file.read())
my_file.close()
# with open("contacts.txt","a+") as f:
    #f.write(page_two_text)
    #f.seek(0)
    #print(f.read())
    #f.close()
```

```python
# CHALLENGE Solution (re-run the %%writefile cell above to obtain an unmodified contacts.txt file):
import re
my_file = open('contacts.txt','a+')
page_two_text = "".join(re.findall(r"[^AUTHORS:]",page_two_text))
my_file.write(page_two_text)
my_file.seek(0)
print(my_file.read())
my_file.close()

#Other solution
# with open("contacts.txt","a+") as f:
#     f.write(page_two_text[8:])
#     f.seek(0)
#     print(f.read())
```

## Regular Expressions
#### 6. Using the `page_two_text` variable created above, extract any email addresses that were contained in the file `Business_Proposal.pdf`.

```python
import re

# Enter your regex pattern here. This may take several tries!
pattern = r"\w*@\D*.com"

re.findall(pattern, page_two_text)

#Other solution : pattern = r"[\w]+@[\w]+.\w{3} [] : set !!
```
