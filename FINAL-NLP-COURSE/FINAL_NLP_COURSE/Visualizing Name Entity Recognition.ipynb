```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
from spacy import displacy
```

```python
doc = nlp(u"Over the last quarter Apple sold nearly 20 thousand iPods for a profit of $6 milion"
         u"By contrast, Sony only sold 8 thousand Walkman music players.")
```

```python
displacy.render(doc,style="ent",jupyter=True)
```

```python
for sent in doc.sents:
    displacy.render(nlp(sent.text),style="ent",jupyter=True)
```

```python
colors = {"ORG":"red"}
options = {"ents":["PRODUCT","ORG"],"colors":colors}
```

```python
displacy.render(doc,style="ent",jupyter=True,options=options)
```

```python
colors = {"ORG":"radial-gradient(yellow,red)"}
options = {"ents":["PRODUCT","ORG"],"colors":colors}
```

```python
displacy.render(doc,style="ent",jupyter=True,options=options)
```

```python
colors = {"ORG":"linear-gradient(90deg,#aa9cfc,red)"}
options = {"ents":["PRODUCT","ORG"],"colors":colors}
```

```python
displacy.render(doc,style="ent",jupyter=True,options=options)
```

```python
displacy.serve(doc,style="ent",options=options,port = 5001)
```
