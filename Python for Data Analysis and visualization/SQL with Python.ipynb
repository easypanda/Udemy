```python
def sql_to_df(sql_query):
    #Use pandas to pass sql query using the sqlite3 connection
    df = pd.read_sql(sql_query,con)
    return df
```

```python
#Select multiple columns example

query = '''SELECT first_name,last_name FROM customer;'''

sql_to_df(query).head()
```

```python
#Select multiple columns example

query = '''SELECT * FROM customer;'''

sql_to_df(query).head()
```

```python
#Select distinct country_ids from the city table

query = '''SELECT DISTINCT(country_id) FROM city;'''

sql_to_df(query).head()
```

```python
#Select all the customers info from the 1st store

query = ''' SELECT *
            FROM customer
            WHERE store_id = 1'''

sql_to_df(query).head()
```

```python
#Select all customer info from Mary

query = ''' SELECT *
            FROM customer
            WHERE first_name = "MARY"'''

sql_to_df(query).head()
```

```python
#Select all films from 2006 and rated R

query = ''' SELECT *
            FROM film
            WHERE release_year = 2006
            AND rating = "R"'''

sql_to_df(query).head()
```

```python
#Select all films from R or PG

query = ''' SELECT *
            FROM film
            WHERE rating = "PG"
            OR rating = "R"'''

sql_to_df(query).head()
```

List of aggregate functions:

*AVG() : AVERAGE

*COUNT() : NUMBER OF ROWS

*FIRST() : FIRST VALUE

*LAST() : LAST VALUE

*MAX() : MAX VALUE

*MIN() : MIN VALUE

*SUM(): SUM VALUE

```python
#Count the number of customers

query = ''' SELECT COUNT(customer_id)
            FROM customer;'''

sql_to_df(query).head()
```

```python
query = ''' SELECT rental_rate,rental_duration,
            AVG(rental_rate),SUM(rental_duration)
            FROM film;'''

sql_to_df(query).head()
```

```python
#SQL widlcard % 

#Select any customers whose name start with M

query = ''' SELECT *
            FROM customer
            WHERE first_name LIKE "M%";'''

sql_to_df(query).head()
```

```python
#SQL widlcard _ 

#Select any customers whose last name ends with ing

query = ''' SELECT *
            FROM customer
            WHERE last_name LIKE "_ing";'''

sql_to_df(query).head()
```

```python
#The [character_list] wildcard.
#BE CAREFUL, the syntax GLOB'[char_list]*' is spectific to SQLite, and would be LIKE '[char_list]%' with MYSQL

#Select any customers whose first name begins with A or B

query = ''' SELECT *
            FROM customer
            WHERE first_name GLOB '[AB]*';'''

sql_to_df(query).head()
```

```python
#Select all customers and order results by last name

query = ''' SELECT *
            FROM customer
            ORDER BY last_name; '''

sql_to_df(query).head()
```

```python
#Select all customers and order results by last name DESCENDING

query = ''' SELECT *
            FROM customer
            ORDER BY last_name DESC; '''

sql_to_df(query).head()
```

```python
#Count the number of customers per store

query = ''' SELECT store_id, COUNT(customer_id)
            FROM customer
            GROUP BY store_id; '''

sql_to_df(query).head()
```
