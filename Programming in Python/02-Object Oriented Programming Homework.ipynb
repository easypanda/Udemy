#### Problem 2

Fill in the class 

```python
class Cylinder:
   
    pi = 3.14
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self):
        return self.height * self.radius * self.radius* Cylinder.pi
    
    def surface_area(self):
        return 2 * Cylinder.pi * self.radius ** 2 + (2 * Cylinder.pi * self.radius * self.height)
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
